```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate some sample data
# True line: y = 3x + 2
X = np.linspace(-10, 10, 100)
Y = 3 * X + 2 + np.random.randn(*X.shape) * 2  # added noise

# Step 2: Build linear model with TensorFlow
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,))   # y = mx + c
])

# Step 3: Compile model
model.compile(optimizer='sgd', loss='mse')

# Step 4: Train model
history = model.fit(X, Y, epochs=200, verbose=0)

# Step 5: Get learned parameters
weights, bias = model.layers[0].get_weights()
print(f"Learned slope (m): {weights[0][0]:.3f}")
print(f"Learned intercept (c): {bias[0]:.3f}")

# Step 6: Predict and visualize
Y_pred = model.predict(X)

plt.scatter(X, Y, label="Data", alpha=0.6)
plt.plot(X, Y_pred, color="red", label="Learned Line")
plt.legend()
plt.show()
```
